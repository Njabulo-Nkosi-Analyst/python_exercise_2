### Importing libraries

```python

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

## Importing data/ data ingesion

```python

#importing data in spark library
data_path = spark.read.table("workspace.default.retail_sales_dataset")


## converting it to pandas
df = data_path.toPandas()

#showing top 10 rows in a dataframe
df.head(10)
```

### EDA

```python
##checking the infomation about my columns

df.info()
```

```python
#converting date column to datetime

df['Date']= pd.to_datetime(df["Date"])

#display
df['Date'].dtype
```

### questions & visualization

```python


#1 Create a Bar Chart: Revenue by product
revenue_by_product =(df.groupby('Product Category',as_index=False).agg(Total_Amount=('Total Amount','sum')).sort_values(by='Total_Amount',ascending=False))
print(revenue_by_product)

# visuals(bar graph)
plt.bar(revenue_by_product['Product Category'], revenue_by_product['Total_Amount'], color='skyblue', edgecolor='black')

# adding labels 
plt.xlabel('Product Category')
plt.ylabel('Total Revenue (R)')
plt.title('Total Revenue by Product Category')

# show the plot
plt.show()


```

```python
# 2 Create a Line Chart: Total revenue over time
revenue_by_time =df.groupby('Date',as_index=False).agg(Total_Amount=('Total Amount','sum'))

#print(revenue_by_time)

#visuals(Line Chart)
plt.bar(revenue_by_time['Date'], revenue_by_time['Total_Amount'], color='blue')

#adding labels 
plt.xlabel('Date')
plt.ylabel('Total Revenue (R)')
plt.title('Total Revenue over time')

#show the plot
plt.show()
```

```python
# 3 Create a Histogram: Distribution of revenue
revenue = df['Total Amount']

# 2. Creating histogram using matplotlib
plt.hist(revenue, bins=10, color='skyblue', edgecolor='black')

#adding labels 
plt.xlabel('Transaction Amount')
plt.ylabel('Frequency (Number of Sales)')
plt.title('Distribution of Sales Amounts')

# 4. Display 
plt.show()
```
